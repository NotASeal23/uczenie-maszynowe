```python
column_names = [
    "pclass", "survived", "name", "sex", "age", "sibsp", "parch",
    "ticket", "fare", "cabin", "embarked", "boat", "body", "home.dest"
]

df = pd.read_csv(
    "titanic.csv",
    names=column_names,
    na_values="?"
)
```

Dokonuję redukcję tej zmiennej, ponieważ różne wartości są inne ale można je pogrupować na podstawie podobieństwa. Pierwsza litera mówi prawdopodobnie o sektorze. Dzięki redukcji nie musimy bawić się z poszczególnymi pokojami osobno, ale możemy wziąć sektory jako całość. Niestety z tą operacją wiąże się utrata dokładnych, szcegółowych danych o lokalizacji. Może to mieć wpływ, szczególnie gdy znamy schemat statku (np. najdalsze pokoje w korytarzu skutkowały trudnośćią w wydostaniu się ze statku).
